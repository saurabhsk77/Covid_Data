```python
from covid import Covid
import matplotlib.pyplot as plt

cov=Covid()

countryName=input('Enter the country name ')

covidData=cov.get_status_by_country_name(countryName)
confirmed=covidData['confirmed']
active=covidData['active']
recover=covidData['recovered']
deaths=covidData['deaths']

print("CONFIRMED CASES-",confirmed)
print("ACTIVE CASES-",active)
print("RECOVERED CASES-",recover)
print("DEATH-",deaths)


explode=(0.1,0.1,0.1,0.1)
plt.pie([confirmed,active,recover,deaths],labels=['confirmed','active','recover','death'],explode=explode,shadow=True,startangle=20)
plt.title(countryName)
plt.legend()
plt.show()
```
